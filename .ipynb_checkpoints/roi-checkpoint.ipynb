```python
from IPython.display import clear_output
import webbrowser
import time
class income:
    def __init__(self):
        self.rent = 0
        self.laundry = 0
        self.storage = 0
        self.misc = 0

    def total(self):
        global total_income 
        total_income = 0
        total_income += int(self.rent)
        total_income += int(self.laundry)
        total_income += int(self.storage)
        total_income += int(self.misc)
        self.income_var = total_income
        return total_income

    def rental_income(self):
        clear_output()
        self.total()
        print("""Please enter whole dollar amounts. Avoid using any '$' or commas.""")
        self.rent = input("How much is your monthly rental income?: $")
        clear_output()
        print(f"*Total monthly income: ${self.rent}*")
        self.extras = input("Do you have income from any of the following?: Laundry, storage, misc. Enter 'Yes' or 'No': ").title()
        while self.extras not in ("Yes", "No"):
            self.extras = input("Invalid answer, Please enter yes or No.").title()
        if self.extras == "Yes":
            clear_output()
            self.laundry = input("What is your monthly laundry service income?: $")
            clear_output()
            print(f"*Total monthly income: ${int(self.rent) + int(self.laundry)}*")
            self.storage = input("What is your monthly storage service income?: $")
            clear_output()
            print(f"*Total monthly income: ${int(self.rent) + int(self.laundry) + int(self.storage)}*")
            self.misc = input("Please enter total amount of any other income: $")
            clear_output()
            self.total()
        else:
            clear_output()
            self.total()
        print(f"""
        \t\t\t\tRental income: ${self.rent}
        \t\t\t\tLaundry income: ${self.laundry}
        \t\t\t\tStorage income: ${self.storage}
        \t\t\t\tMisc. income: ${self.misc}""")
        print(100*"=")
        print(f"\t\t\t\t\tTotal: ${self.income_var}")
        print(100*"=")
        self.confirm = input("Is this information correct? Enter 'Yes' or 'No'").title()
        if self.confirm == "No":
            total_income = 0
            self.rental_income()
        else:
            print("Added")
    
class expenses:

    def __init__(self):
        self.tax = 0
        self.insurance = 0
        self.utilities = 0
        self.hoa = 0
        self.vaccancy = 0
        self.repairs = 0
        self.capex = 0
        self.manage = 0
        self.mortgage = 0
        
    def total(self):
        global total_expense
        total_expense = 0
        total_expense += int(self.tax)
        total_expense += int(self.insurance)
        total_expense += int(self.utilities)
        total_expense += int(self.hoa)
        total_expense += int(self.vaccancy)
        total_expense += int(self.repairs)
        total_expense += int(self.capex)
        total_expense += int(self.manage)
        total_expense += int(self.mortgage)
        self.expense_var = total_expense
        return total_expense
    def total_expenses(self):
        clear_output()
        self.total()
        print("""Please enter whole dollar amounts. Avoid using any '$' or commas.""")
        self.tax = input("How much are you monthly taxes on property? If you dont know enter 'Help': ").title()
        if self.tax == "Help":
                tax_link = webbrowser.open("https://smartasset.com/taxes/property-taxes#5v5qVaLtbB")
                tax_divide = input("Enter the annual tax calculation the website provided: $")
                self.tax = int(tax_divide) // 12
        clear_output()
        print(f"${self.tax} added")
        self.insurance = input("How much is your monthly insurance? (The average cost of homeowners insurance is $1,249 per year, or $104 per month): $")
        clear_output()
        print(f"${self.insurance} added")
        if_utilities = input("Who's paying utilities(electric, gas, etc.) 1. Owner(You) 2. Tenants, enter 1-2")
        if if_utilities == '1':
            self.util()
            self.utilities = self.utility_total
        clear_output()
        print(f"${self.utilities} added")
        self.hoa = int(input("What's your HOA fees?: $"))
        clear_output()
        print(f"${self.hoa} added")
        self.vaccancy = input ("Would you like to set money aside incase of vaccancy? This is highly reccomended. Reccomended amount is 5% of total income. To use reccomended amount enter 'Yes', other wise enter dollar amount: $").title()
        if self.vaccancy == "Yes":
      
            self.vaccancy = (total_income *.05)//1
        clear_output()
        print(f"${self.vaccancy} added")
        self.repairs= input("It is reccomended to set aside money for repairs ($50-$100). How much would you like to set aside for repairs? : $")
        clear_output()
        print(f"${self.repairs} added")
        self.capex = input("It is reccomended to set aside money for capital expenditures AKA CapEx ($50-$100). How much would you like to set aside for CapEx?: $")
        clear_output()
        print(f"${self.capex} added")
        prop_manage = (total_income * .10)
        self.manage = input(f"If you will be handling your own repairs and upkeep, enter 0. Otherwise enter how much you will be paying for property management. For your property, this is esitmated at ${prop_manage} (10% of total income).: $")
        clear_output()
        print(f"${self.manage} added")
        self.mortgage = input("How much is your mortgage? If not sure enter 'Help' and you will be directed to a mortgage calculator: $").title()
        if self.mortgage == 'Help':
            webbrowser.open("https://www.google.com/search?q=mortgage+calculator&rlz=1C1CHZN_enUS977US981&oq=mortgage+&aqs=chrome.0.0i131i433i512j0i433i512j69i57j0i433i512j0i512j0i433i512l2j0i512l3.567270j0j15&sourceid=chrome&ie=UTF-8")
            self.mortgage = input("Enter how much monthly mortgage is: $")
        self.total()
        clear_output()
        print(f"""
                            Insurance: ${self.insurance}
                            Utilities: ${self.utilities}
                            HOA: ${self.hoa}
                            Vaccany: ${self.vaccancy}
                            Repairs: ${self.repairs}
                            Capex: ${self.capex}
                            Property Management: ${self.manage}
                            Mortgage: ${self.mortgage} """)
                        
        print("="*100)
        print(f"\t\t\tYour final expense total is ${total_expense}")
        print("="*100)
        confirm = input("Is this information correct? Enter 'Yes' or 'No': ")
        if confirm == 'No':
            self.total_expenses()

    def util(self):
        clear_output()
        self.electricity = int(input("How much is electricity monthly?: $"))
        self.water = int(input("How much is water monthly?: $"))
        self.gas = int(input("How much is gas monthly?: $"))
        self.sewage = int(input("How much is sewage monthly?: $"))
        self.garbage = int(input("How much is garbage monthly?: $"))
        clear_output()
        self.util_total()
        print(f"""
        \t\t\t\tElectricity: ${self.electricity}
        \t\t\t\tWater: ${self.water}
        \t\t\t\tGas: ${self.gas}
        \t\t\t\tSewage: ${self.sewage}
        \t\t\t\tGarbage: ${self.garbage}""")
        print(100*"=")
        print(f"\t\t\t\t\t total: ${self.utility_total}")
        print(100*"=")
        confirm = input("Is the following information correct? enter 'Yes' or 'No': ").title()
        if confirm == "No":
            self.util()
        else:
            clear_output()
    def util_total(self):
        self.utility_total = 0
        self.utility_total += self.electricity
        self.utility_total += self.water
        self.utility_total += self.gas
        self.utility_total += self.sewage
        self.utility_total += self.garbage

class cash_flow:
    def cash_info(self):
        clear_output()
        print("Calculating cash flow...")
        time.sleep(3)
        self.cash_total()
        clear_output()
        print(f""" 
        Your monthly cash flow is ${flow_total} for an annual amount of ${(flow_total*12)}. This is calculated by 
        subtracting your expenses from your total monthly income. Enter 'ok' to continue to R.O.I.
        or 'Help' to redo income or expenses.
        *Total income: ${total_income}   *Total expenses: ${total_expense}
        """)
        confirm = input("Enter here: ").title()
        if confirm == "Help":
            edit = input("If you would like to redo income enter '1'. If you would like to redo expenses enter '2'.")
            if edit == '1':
                start= income()
                start.rental_income()
                self.cash_info()
            elif edit =='2':
                next1 = expenses()
                next1.total_expenses()
                self.cash_info()
        while confirm not in ('Help', 'Ok'):
            confirm = input("Invalid response, please enter 'Ok' or 'Help'").title()
    def cash_total(self):
        global flow_total
        flow_total = total_income - total_expense
        
class ROI:
    def __init__(self):
        self.down_payment = 0
        self.closing_cost = 0
        self.renovate = 0
        self.misc = 0
    def roi_total(self):
        clear_output()
        print("""Please enter whole dollar amounts. Avoid using any '$' or commas.""")
        self.down_payment = input("How much is the down payment?: $")
        clear_output()
        print(f"Down payment of ${self.down_payment} added")
        self.closing_cost = input("How much is the closing cost?: $")
        clear_output()
        print(f"Closing cost of ${self.closing_cost} added")
        self.renovate = input("How much will you use to renovate the place?: $")
        clear_output()
        print(f"Renovation budget of ${self.renovate} added")
        self.misc = input("Amount of any other investments in the house: $")
        clear_output()
        self.investment_total()
        print(f"""
                                Down Payment: ${self.down_payment}
                                Closing Cost: ${self.closing_cost}
                                Renovation: ${self.renovate}
                                Misc: ${self.renovate}
        """)
                                
        print("="*100)
        print(f"""         \t\tYour final investment total is ${self.total_investment}""")
        print("="*100)
        confirm = input("\t\tIs this information correct? Enter 'Yes' or 'No': ").title()
        if confirm == "No":
            self.roi_total()
        self.calc_roi()
        clear_output()
        print("Calculating...")
        time.sleep(3)
        clear_output()
        print(f"Your R.O.I. is {round(self.roi,2)}%")
        again = input("Would you like to calculate another property? Enter 'Yes' for main menu or 'No' to exit").title()
        if again == "Yes":
            roi_calc()
            
        elif again == "No":
            clear_output()
            return
    def investment_total(self):
        self.total_investment = 0
        self.total_investment += int(self.down_payment)
        self.total_investment += int(self.closing_cost)
        self.total_investment += int(self.renovate)
        self.total_investment += int(self.misc)
        return self.total_investment
    def calc_roi(self):
        self.annual_cash = flow_total * 12
        self.roi = self.annual_cash / self.total_investment
        
def roi_calc():
    start = income()
    next1 = expenses()
    next2= cash_flow()
    end = ROI()
    clear_output()
    print(100*"=")
    print("""       \t\tWelcome to the ROI calculator. Making Long term Investment Easy""")
    print(100*"=")
    
    choice = input("To begin enter '1' or to exit enter '2'.")
    if choice == '1':
        start.rental_income()
        next1.total_expenses()
        next2.cash_info()
        end.roi_total()
    elif choice == "2":
        clear_output()
        return

    else:
        choice= input("Invalid choice. Please enter 1-2")

roi_calc()
```
